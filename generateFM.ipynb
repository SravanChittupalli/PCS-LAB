```python
import scipy.signal as signal
import numpy as np
import matplotlib.pyplot as plt
```

```python

fc = int(input("Enter frequency of carrier: "))
deviation_ratio = float(input("Enter deviation ratio: "))
kf = int(input("Enter kf value: "))
```

```python
fs= 6000 # sampling freq
T= 1    # TimePeriod in sec
twopi= 2*np.pi 
```

```python
t = np.linspace(-T , T, num=fs, endpoint=False)
t
```

```python
message_func = []
f = open('data/SampleECG_converted.txt', 'r')
lines = f.readlines()
for line in lines:
    message_func.append(float(line.split(' ')[1])*kf)
message = message_func
```

```python
carrier = np.cos(twopi*fc*t)
carrier
```

```python
output = np.cos(twopi*fc*t + message)
output
```

```python
plt.subplot(3,1,1)
plt.title("Message Signal")
plt.plot(t,message,'green')
plt.plot(t , 0*t,"black")
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
plt.subplot(3,1,1)
plt.title("Carrier Signal")
plt.plot(t,carrier,'red')
plt.plot(t , 0*t,"black")
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
plt.subplot(3,1,1)
plt.title("FM Signal")
plt.plot(t,output,'blue')
plt.plot(t , 0*t,"black")
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

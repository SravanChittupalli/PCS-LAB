## Generating FM Wave
PCS Lab1

```python
import scipy.signal as signal
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs= 10000 # sampling freq
fc= 10   # carrier freq
fm= 1   # modulation freq
```

```python
T= 10    # TimePeriod in sec
twopi= 2*np.pi 
beta = 100
```

```python
t = np.linspace(0 , T, num=fs, endpoint=False)
t
```

```python
message = np.cos(twopi*fm*t)
message
```

```python
carrier = np.cos(twopi*fc*t)
carrier
```

```python
output = np.cos(10*t + beta*np.cos(1*t))
output
```

```python
plt.subplot(3,1,1)
plt.title("Message Signal")
plt.plot(message,'green')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
plt.subplot(3,1,1)
plt.title("Carrier Signal")
plt.plot(carrier,'red')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
plt.subplot(3,1,1)
plt.title("FM Signal")
plt.plot(output,'blue')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

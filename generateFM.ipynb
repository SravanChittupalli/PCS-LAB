## Generating FM Wave

```python
import scipy.signal as signal
import numpy as np
import matplotlib.pyplot as plt
```

```python
message_func = input("Enter message signal equation: ")
fc = int(input("Enter frequency of carrier: "))
deviation_ratio = float(input("Enter deviation ratio: "))
kf = int(input("Enter kf value: "))
```

```python
fs= 400000 # sampling freq
T= 1    # TimePeriod in sec
twopi= 2*np.pi 
```

```python
t = np.linspace(-T , T, num=fs, endpoint=False)
t
```

```python
message = eval("np."+message_func)
message
```

```python
carrier = np.cos(twopi*fc*t)
carrier
```

```python
output = np.cos(twopi*fc*t + kf*message)
output
```

```python
plt.subplot(3,1,1)
plt.title("Message Signal")
plt.plot(t,message,'green')
plt.plot(t , 0*t,"black")
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
plt.subplot(3,1,1)
plt.title("Carrier Signal")
plt.plot(t,carrier,'red')
plt.plot(t , 0*t,"black")
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
plt.subplot(3,1,1)
plt.title("FM Signal")
plt.plot(t,output,'blue')
plt.plot(t , 0*t,"black")
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

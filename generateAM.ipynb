```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
message_func = input("Enter message signal equation: ")
```

```python
fs= 4000
fc= 10
fm= 1
T = 10
```

```python
twopi= 2*np.pi
t = np.linspace(-T , T, num=fs, endpoint=False)
t
```

```python
carrier = np.cos(twopi*fc*t)
message = eval("np."+message_func)
```

```python
plt.subplot(3,1,1)
plt.title("Message Signal")
plt.plot(message,'green')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
plt.subplot(3,1,1)
plt.title("Message Signal")
plt.plot(carrier,'green')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
output = (3 + message)*carrier
```

```python
plt.subplot(3,1,1)
plt.title("Message Signal")
plt.plot(output,'green')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

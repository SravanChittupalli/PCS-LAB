```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

mod_index = float(input("Enter required modulating index: "))
fc = int(input("Enter carrier frequency: "))
```

```python
message_func = []
f = open('data/SampleECG_converted.txt', 'r')
lines = f.readlines()
for line in lines:
    message_func.append(float(line.split(' ')[1]))
```

```python
fs= 6000   # Sampling frequency
T = 1
```

```python
twopi= 2*np.pi
t = np.linspace(-T , T, num=6000, endpoint=False)
t
```

```python

amplitude_carier = -(min(message_func))/mod_index
carrier = np.cos(twopi*fc*t)
message = message_func
print(min(message_func) , amplitude_carier)
```

```python
plt.subplot(3,1,1)
plt.title("Message Signal")
plt.plot(t,message,'green')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
plt.subplot(3,1,1)
plt.title("Carrier Signal")
plt.plot(t,carrier,'green')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
output = amplitude_carier*carrier + message*carrier
```

```python
plt.subplot(3,1,1)
plt.title("AM Signal")
plt.plot(t,output,'green')
plt.plot(t,0*t,"black")
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

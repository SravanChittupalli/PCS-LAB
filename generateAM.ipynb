```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
message_func = input("Enter message signal equation: ")
mod_index = int(input("Enter required modulating index: "))
fc = int(input("Enter carrier frequency: "))
```

```python
fs= 40000   # Sampling frequency
T = 1
```

```python
twopi= 2*np.pi
t = np.linspace(-T , T, num=fs, endpoint=False)
t
```

```python
carrier = np.cos(twopi*fc*t)
message = eval("np."+message_func)
```

```python
plt.subplot(3,1,1)
plt.title("Message Signal")
plt.plot(t,message,'green')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
plt.subplot(3,1,1)
plt.title("Carrier Signal")
plt.plot(t,carrier,'green')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
output = (3 + message)*carrier
```

```python
plt.subplot(3,1,1)
plt.title("AM Signal")
plt.plot(t,output,'green')
plt.plot(t,0*t,"black")
plt.xlabel('Time')
plt.ylabel('Amplitude')
```
